```python
songids=[]
lyrics=[]
elbumids=[]
genres=[]
a=datetime.datetime.now()
```

```python
chrome_options=Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--window-size=1920x1080")
driver=Chrome(options=chrome_options)
```

```python
for k in rsong['songid']:
    time.sleep(random.random())
    print('https://www.melon.com/song/detail.htm?songId='+str(k))
    try:
        driver.get('https://www.melon.com/song/detail.htm?songId='+str(k))
        bs = BeautifulSoup(driver.page_source,'html.parser')
        songids.append(k)
        try:
            lyrics.append(bs.select('#d_video_summary')[0].text.strip())
        except IndexError:
            lyrics.append('None')
        elbumids.append(bs.select('.list > dd > a')[0].get('href').split('\'')[-2])
        genres.append(bs.select('.list > dd')[2].text)
    except UnexpectedAlertPresentException:
        print('error')
        time.sleep(5)
        next
    if len(songids)%100 ==0:
        print(str(len(songids))+'/'+str(wh))
        time.sleep(random.random()*8)
print(datetime.datetime.now()-a)
```

```python
df=pd.DataFrame({'songid':songids,
                'elbumids':elbumids,
                'genres':genres,
                'lyrics':lyrics})
df
```
